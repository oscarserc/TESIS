#**<font color='brown'>· <u>Introducción</u>.</font>**
Vivimos en un mundo cada vez más interconectado, donde las comunicaciones juegan un papel de creciente importancia. Esto hecho posibilita que la influencia de unos individuos sobre otros se difunda más facilmente.
Como consecuencia de ello nuestra manera de comportarnos, o la forma de interpretar los acontecimientos que nos rodean puede verse alterada por
individuos con los que entramos en contacto.

Conscientes de ello, todo tipo de actores maliciosos han trasladado su actividad a la red con la finalidad de conseguir sus propios fines.

La proliferación, por ejemplo, de fenómenos como la difusión de noticias falsas o la creación interesada de estados artificiales de opinión, son un reflejo de los problemas a los que nos enfrentamos. Tal y como alertan algunos estudios, nos encontramos cada vez con más frecuencia con la utilización incluso de medios automatizados (*bots* ) para influir en el clima de opinión en las redes sociales on-line, y con ello polarizar y alterar el equilibrio de la red social <font color='blue'> [Ferrara, E. et al. (2017)] </font>.
En una sociedad avanzada, en la que la opinión pública juega un importante papel, este tipo de manipulaciones pueden generar un peligroso efecto de desestabilización.

Otro claro ejemplo de actividad maliciosa lo encontramos en el uso de las redes para acosar a otras personas, algo que es todavía más grave para ciertos sectores de la población especialmente sensibles. Los adolescentes, por ejemplo, se encuentran ante una creciente exposición en las redes sociales on-line. Según revelan algunos estudios alrededor de un 15% de ellos sufren acoso en ellas <font color="blue">[Young, E. (2024)]</font>.
En contra de lo que pueda parecer, lejos de suponer una menor amenaza, las interacciones on-line activan iguales patrones mentales que en las relaciones en el mundo real, y suponen un efecto directo en la generación de estrés emocional <font color="blue">[Nesin, S. et al (2025)]</font>. El <b>anonimato</b> es un factor adicional a tener en cuenta en estos casos, ya que puede acabar conduciendo a experiencias muy traumáticas por la indefensión e impotencia de la persona que lo sufre <font color="blue">[Nixon, C.L. (2014)]</font>. En situaciones como éstas, el temor y la vergüenza de la víctima pueden acabar silenciando e invisibilizando el problema, lo que dificulta su estudio y análisis, así como la toma de medidas orientadas a su solución.

Creemos firmemente que la modelización matemática de este tipo de dinámicas destructivas puede contribuir decisivamente a contener la conflictividad en una red social, así como ayudar en la toma de medidas que faciliten el acabar con este tipo de prácticas abusivas.

De todos los fenómenos mencionados nos centraremos en el tema del acoso en una red social. Dada la vulnerabilidad de los sectores de la población que se ven afectados tiene especial interés el dedicar la Tesis Doctoral a su estudio.

**A) <u>Simulación de una Red Social</u>**.

El primer desafío al que nos enfrentamos es el de simular el funcionamiento de una Red Social. La representación de la actividad de la red es el reto inicial que debemos resolver en este sentido.

Lo más adecuado podría consistir en representar la actividad de la red registrando el estado de cada agente en cada momento, y para ello emplearíamos algún tipo de propiedad que se pueda ligar a las interacciones que tiene con los agentes con los que entra en contacto. Una posible respuesta podría consistir en la utilización de <b>emociones</b>. Muchas de las redes sociales (<i>Twitter, China Weibo, entre otras</i> ) tienen alguna forma de etiquetado emocional del usuario, y existen técnicas de análisis de sentimientos que permiten lograr así mismo dicho objetivo.

El siguiente reto consistirá en encontrar alguna vía de generar el estado emocional de los agentes. Podríamos emplear para ello, algún tipo de juego como mecanismo de interacción. Las distintas situaciones que podrían aparecer, según la acción de cada agente, servirían como desencadenante de la respuesta emocional (<i>sentirse correspondido induciría alegría, verse traicionado provocaría ira, etc</i> ).

La rama matemática conocida como **Teoría de Juegos** proporciona abundantes esquemas formalizados de interacción de este tipo que podrían servirnos. De entre todos ellos hemos elegido **<i>El Dilema del Prisionero</i>**, tal y como analizaremos posteriormente. Los agentes, además, podrían guiarse por algún tipo de estrategia particular, como puede ser la de acosar y aprovecharse de los demás sistemáticamente. Nuestro objetivo consistirá en identificar la presencia de este tipo de jugadores y su localización en la red.

#**<font color='brown'>· <u>Fase I: SIMULACIÓN de la Red Social</u>.</font>**

Si observamos el funcionamiento de las redes sociales, vemos que en muchas de ellas, tal y como se ha comentado previamente, se etiqueta (<i>o es posible hacerlo</i> ) el estado emocional del usuario. Por ello esta es una vía mediante la que es posible codificar el estado de los agentes en una red social.

Como mecanismo que tenga cierta complejidad y que pueda abarcar algunas de las principales reacciones emocionales, hemos elegido un juego de los planteados por la Teoría de Juegos: **El Dilema del Prisionero**. Mediante la dinámica de juego que se establezca entre los agentes de la red, se generará su estado emocional que es lo que registraremos y codificaremos como una secuencia de imágenes.

En lo que sigue analizamos algunos de los resultados existentes en la literatura que nos permiten diseñar, de manera más precisa, un conjunto de agentes que se interrelacionan en una red social, y en donde su conducta se encuentra muy influenciada por las emociones.

Nos centraremos inicialmente en la forma en que las emociones pueden servir de mecanismo de difusión, de influencia entre los individuos que operan en la red social.

<br></br>

#**1. MECANISMO DE DIFUSIÓN**
#**Influencia de las EMOCIONES en las interacciones en una Red Social**.

Cuando analizamos el comportamiento de individuos en una Red Social se observan fenómenos llamativos en los cuales las emociones parecen jugar un papel fundamental. Si queremos simular una Red Social, y que las emociones jueguen un papel esencial, tendremos que reflejar este tipo de rasgos.

<b>EFECTO LOCAL</b>

Algunos estudios como el de Framingam, muestran el contagio de las emociones por la red y la creación de clusters: los individuos suelen mostrar, incluso a largo plazo, estados emocionales sintonizados con su vecindario más cercano (<i>ver fig. 4</i> ) <font color='blue'> [Fowler, J. H. & Christakis, N. A. (2008)] </font>.

- Las emociones nos atan a las personas que nos rodean y han de desencadenarse a través de las interacciones que tenemos con ellas. Este es un efecto local.

<center>

<b>EFECTO MACROSCÓPICO</b>

Las emociones, por otra parte, parecen impulsar la viralización de información <font color='blue'> [Berger, J. \& Milkman, K. (2011)]</font>. La propagación de sentimientos en redes sociales se ha detectado en diversos medios como Livejournal <font color="blue">[Zafarani, R. et al. (2010)]</font>, o en Twitter <font color="blue">[Romero, D. M. y et al. (2011)]</font>. También aparecen evidencias experimentales de contagios emocionales en gran escala en redes sociales <font color="blue">[Kramer, A. D. I. et al. (2014)]</font> y a través de Facebook <font color="blue">[Kramer, A. D. I. (2012)]</font>. Por tanto:


- El funcionamiento de la simulación de la red social ha de poder desencadenar un efecto macroscópico de contagio.


<br></br>
La simulación que generemos de la Red Social ha de reflejar estos dos aspectos que se detectan en las Redes Sociales. Por un lado interacciones locales que se manifiestan localmente en el estado emocional de los nodos. Y por otro lado esta manifestación local ha de poder transformarse en una difusión que genere un efecto macroscópico visible de contagio emocional por la red.

El factor que falta por añadir es la conexión entre las emociones y las acciones que llevamos a cabo. Según <font color="blue">[Rolls, Edmund T. (2014)]</font> las emociones vendrían a ser un mecanismo de refuerzo que interviene en el aprendizaje instrumental, lo que nos permite desarrollar acciones adaptadas para optimizar nuestra actuación de cara a solucionar los problemas a los que nos enfrentamos.

Este enfoque estratégico nos sugiere centrar la atención en la Teoría de Juegos para diseñar el mecanismo de interacción de la red.

<br></br>

<b>Toma de decisiones y emociones.</b>

Los trabajos de algunos autores han puesto de manifiesto el papel que tienen las emociones en la toma de decisiones y en la forma de comportarnos <font color='blue'>[Damasio, A.R. (1994)]</font>. Por ello es natural plantearse la posibilidad de que las emociones - con su capacidad de contagio y propagación viral tal y como hemos comentado anteriormente - sean capaces de propiciar la extensión de una determinada conducta (por ejemplo favorecer el surgimiento de estrategias cooperativas).

En <b>Teoría de Juegos</b> ésta es una cuestión que se ha estudiado extensamente. En concreto existen abundantes estudios sobre cómo emerge la cooperación y se propaga como estrategia en juegos repetidos (e.g. <font color="blue">[Axelrod, R. \& Hamilton, W. D. (1981)]</font>, o <font color="blue">[Nowak, M. A. (2006)]</font>). Esto sugiere la idea de emplear mecanismos de interacción como los que plantea esta rama de las matemáticas y economía, como base el funcionamiento de la simulación.

Es por ello que a las anteriores características tengamos que añadir que:
- Las emociones han de dirigir en mayor o menor medida las acciones de los agentes.

Teniendo en cuenta que si la información que disponemos de las relaciones entre individuos es su estado emocional, podemos emplear la perturbación emocional que provocan los nodos conflictivos para detectar su presencia y localización en la red. Y las emociones serán generadas por un juego que practican los individuos de manera iterada.

#**2. Objetivos a alcanzar**:

A partir del estado emocional de los agentes de una Red Social, nos proponemos encontrar los nodos que están perturbando la red con acciones maliciosas como el **acoso**. La detección de la presencia de estos agentes, así como su localización en la Red Social es esencial.

Esto se concreta en dos OBJETIVOS:

#· SEGMENTACIÓN DE IMÁGENES

Pretendemos detectar los tipos de agentes que puedan existir en la Red Social. La utilización de técnicas de segmentación de imágenes permite que el algoritmo funcione como un clasificador de todos y cada uno de los pixeles de la imagen (nodos de la red social). En la implementación que hemos desarrollado, el <b>target</b> utiliza un array con la identidad de todos los agentes.

<figure>
<center>
<img src='https://github.com/oscarserc/IMAGENES/blob/main/Clasificacion_2a_segm.png?raw=true' width="450"/>
<figcaption>fig. 18: <b>Problema de segmentación</b>. <i>A la izquierda tenemos una imagen de la actividad emocional de la red. Interesa clasificar a todos y cada uno de los nodos (pixeles). La identidad que se oculta detrás de cada nodo es el tipo de predicción que devolvería la Arquitectura Neuronal entrenada. A la derecha tendríamos un posible resultado: '<b>0</b>' -> <i>nodos emocionales</i>, '<b>1</b>' -> <i>nodos mutantes agresores</i></i>.</figcaption></center>
</figure>

<br></br>

Segmentar una imagen (ver fig. 18) es en realidad un problema de clasificación en el que han de ser clasificados todos y cada uno de los pixeles según el elemento al que pertenecen.

Si tenemos en cuenta que estamos representando la actividad de una Red Social mediante imágenes, podemos plantearnos si es posible traducir esta información en la detección de los tipos de agentes de la Red. Para esta tarea tendremos que disponer como <b>target</b> del grupo de pertenencia de cada agente (ver fig. 19). Esta información tendremos que guardarla para guiar el proceso de aprendizaje. Como se ha comentado la matriz (array bidimensional) se transforma en un array de 1 sóla dimensión, para poder asociarlo a una única obsrvación.
