### Llevamos a cabo un conjunto de ajustes de cada modelo (en concreto 20) para comprobar la ACCURACY para la evolución de la Red Social. Se calcula el valor medio y la dispersión (medida a través de la desviación típica), asi podemos hacernos una idea de si los valores que genera de muestra en muestras varían mucho.

<b><font color="green">Red Neuronal Convolucional   (<i>Convolutional Neural Network</i> - CNN)</font></b>

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos los puntos de las x
x             = [1000, 2000, 3000, 4000]

# CNN
ciclos_1000 = pd.read_csv('accuracy_CNN_20FITs_1000ciclos_4agentes.csv')
ciclos_2000 = pd.read_csv('accuracy_CNN_20FITs_2000ciclos_4agentes.csv')
ciclos_3000 = pd.read_csv('accuracy_CNN_20FITs_3000ciclos_4agentes.csv')
ciclos_4000 = pd.read_csv('accuracy_CNN_20FITs_4000ciclos_4agentes.csv')
y_CNN         = np.mean(np.array([ciclos_1000, ciclos_2000, ciclos_3000, ciclos_4000]), axis=1)
err_CNN       = np.std(np.array([ciclos_1000, ciclos_2000, ciclos_3000, ciclos_4000]), axis=1)

# Extraemos los valores de los np.array
# Utilizamos (1·sigma) para hacernos una idea de la variabilidad
y_CNN = y_CNN.reshape(1,-1).flatten().tolist()
err_CNN = err_CNN.reshape(1,-1).flatten().tolist()
# Intervalo_error:  (media + 2*sigma) - (media - 2*sigma ) = 4 * sigma
err_CNN = [(4.0 * e) for e in err_CNN]

# Dibujamos las lineas
fig, ax = plt.subplots()
ax.set_ylim([0.0, 1.0])
ax.errorbar(x, y_CNN,    yerr= err_CNN    , fmt='o--', label = "CNN")
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.set_xlabel('\n' + 'Ciclos de Entrenamiento' + '\n' + 'de los agentes inteligentes')
ax.set_ylabel('Accuracy (%)')
ax.set_title('ACCURACY de la Red Neuronal Convolucional (CNN)' + '\n')
plt.show()
```

<b><font color="green">Red Neuronal Convolucional 3D </font></b>

```python

# CNN 3D
ciclos_1000 = pd.read_csv('accuracy_CNN3D_20FITs_1000ciclos_4agentes.csv')
ciclos_2000 = pd.read_csv('accuracy_CNN3D_20FITs_2000ciclos_4agentes.csv')
ciclos_3000 = pd.read_csv('accuracy_CNN3D_20FITs_3000ciclos_4agentes.csv')
ciclos_4000 = pd.read_csv('accuracy_CNN3D_20FITs_4000ciclos_4agentes.csv')
y_CNN_3D    = np.mean(np.array([ciclos_1000, ciclos_2000, ciclos_3000, ciclos_4000]), axis=1)
err_CNN_3D  = np.std(np.array([ciclos_1000, ciclos_2000, ciclos_3000, ciclos_4000]), axis=1)

y_CNN_3D = y_CNN_3D.reshape(1,-1).flatten().tolist()
err_CNN_3D = err_CNN_3D.reshape(1,-1).flatten().tolist()
# Intervalo_error:  (media + 2*sigma) - (media - 2*sigma ) = 4 * sigma
err_CNN_3D = [(4.0 * e) for e in err_CNN_3D]

# Dibujamos las lineas
fig, ax = plt.subplots()
ax.set_ylim([0.0, 1.0])
ax.errorbar(x, y_CNN_3D, yerr= err_CNN_3D , fmt='o--', label = "CNN 3D")
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.set_xlabel('\n' + 'Ciclos de Entrenamiento' + '\n' + 'de los agentes inteligentes')
ax.set_ylabel('Accuracy (%)')
ax.set_title('ACCURACY de la Red Convolucional 3D' + '\n')
plt.show()
```

<b><font color="green">Arquitectura U-Net  </font></b>

```python

# CNN UNet
ciclos_1000 = pd.read_csv('accuracy_UNet_20FITs_1000ciclos_4agentes.csv')
ciclos_2000 = pd.read_csv('accuracy_UNet_20FITs_2000ciclos_4agentes.csv')
ciclos_3000 = pd.read_csv('accuracy_UNet_20FITs_3000ciclos_4agentes.csv')
ciclos_4000 = pd.read_csv('accuracy_UNet_20FITs_4000ciclos_4agentes.csv')
y_UNet      = np.mean(np.array([ciclos_1000, ciclos_2000, ciclos_3000, ciclos_4000]), axis=1)
err_UNet    = np.std(np.array([ciclos_1000, ciclos_2000, ciclos_3000, ciclos_4000]), axis=1)

y_UNet = y_UNet.reshape(1,-1).flatten().tolist()
err_UNet = err_UNet.reshape(1,-1).flatten().tolist()
# Intervalo_error:  (media + 2*sigma) - (media - 2*sigma ) = 4 * sigma
err_UNet = [(4.0 * e) for e in err_UNet]

# Dibujamos las lineas
fig, ax = plt.subplots()
ax.set_ylim([0.0, 1.0])
ax.errorbar(x, y_UNet,   yerr= err_UNet   , fmt='o--', label = "UNet")
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.set_xlabel('\n' + 'Ciclos de Entrenamiento' + '\n' + 'de los agentes inteligentes')
ax.set_ylabel('Accuracy (%)')
ax.set_title('ACCURACY de la Arquitectura UNet' + '\n')
plt.show()
```
